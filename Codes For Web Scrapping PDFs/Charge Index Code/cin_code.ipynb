```python
import pandas as pd

# Set the path to the Excel file
excel_file1 = "Company BSE NSE Scrips and Codes.xlsx"

# Set the column names to be used as keys
key_column1 = "isin_code_equity"
key_column2 = "cin_code"

# Read the Excel file into a pandas DataFrame
df1 = pd.read_excel(excel_file1)

# Create a dictionary mapping between the two columns
mapping = dict(zip(df1[key_column1], df1[key_column2]))

# Print the mapping
# print(mapping)
excel_file2="nifty_companies.xlsx"
df2 = pd.read_excel(excel_file2)
cin_code=[]
for a in df2["ISIN Code"]:
    cin_code.append(mapping.get(a))
print(cin_code)
df2["cin_code"]=cin_code
writer = pd.ExcelWriter(excel_file2)
df2.to_excel(writer, index=False)
writer.save()



```
