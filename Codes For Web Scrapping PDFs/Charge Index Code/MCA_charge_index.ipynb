Installing the necessary libraries

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import numpy as np
```

Reading the file containing Companies' Data

```python
df1=pd.read_excel('Company BSE NSE Scrips and Codes.xlsx') 
```

Extracting cin_codes and Company names from the dataframe

```python

companies = df1['cin_code'].tolist() #cin_codes
companies_name=df1['company_name'].tolist() #company names
```

```python
for company in range(0, len(companies) ): #company is an iterator
    if(len(str(companies[company]))==0): # if cin_code is missing, then just go to the next iteration
       continue
    cin_code=str(companies[company])
    print("working on "+companies_name[company]) # running status
    url=f"https://www.zaubacorp.com/companysearchresults/{cin_code}" #url changes for different cin_codes.
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage = requests.get(url, headers = headers) #getting response from the webpage.
    soup = BeautifulSoup( webpage, 'lxml' )
    table = soup.find("table", {"id": "charges"}) #finding the table with the charge index data.
    if( table== None ): # if no table is present, just go to the next iteration
        continue
    table_data = []
    for row in table.find_all('tr'):
        row_data = []
        for cell in row.find_all('td'):
            row_data.append(cell.text.strip())
        table_data.append(row_data) #updating each row of the table.
    df = pd.DataFrame(table_data) #converting the given table into a dataframe.
    filename=companies_name[company] #filename based on the company's name.
    filename=filename+".xlsx"
    df.to_excel(filename, index=False) #saving the file as "companyName.xlsx"
```
